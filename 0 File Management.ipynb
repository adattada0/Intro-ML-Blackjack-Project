# Import Dataset


```python
from google.colab import drive
import pandas as pd
import gc

num_rows = 25000

# Import raw data
drive.mount('/content/drive', force_remount=True)
file_path_raw = '/content/drive/MyDrive/Classes/Intro ML/Project/Dataset/raw.csv'
data = pd.read_csv(file_path_raw, nrows=num_rows)

# Write out new CSV
file_path_processed = '/content/drive/MyDrive/Classes/Intro ML/Project/Dataset/data.csv'
data.to_csv(file_path_processed, index=False)
print(f"Limited data (first {num_rows} rows) saved to {file_path_processed}")

del data, file_path_raw, file_path_processed
gc.collect()
```
